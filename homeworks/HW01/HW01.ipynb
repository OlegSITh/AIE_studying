# 1.Импрот библиотеки Numpy

```python
import numpy as np
```

# 2.Примеры создания массивов

```python
# Одномерный массив
list = [1, 2, 3, 4, 5]
array = np.array(list)

print(list, type(list))
print(array, type(array))
```

```python
# Двумерный массив
twice_d_list = [[1, 2, 3], [4, 5, 6]]
twice_d_array = np.array(twice_d_list)

print(twice_d_list, type(twice_d_list), end="\n\n")
print(twice_d_array, type(twice_d_array))
```

```python
# N-мерный массив
n_d_array = np.full((4,2,5,2,5), 8)

print(n_d_array)
```

```python
# Заполнено нулями
zeros = np.zeros((5,5))
print(zeros)
```

```python
# Заполнено единицами
ones = np.ones((4,4,4))
print(ones)
```

```python
# Заполнено в промежутке
arange = np.arange(0,108)
print(arange)
```

```python
# Заполнено рандомными float
rnd_array = np.random.random(9)
rnd_array_res = rnd_array.reshape(3,3)

print(rnd_array, end="\n\n") # просто 9 элементов
print(rnd_array_res) # перестроены в форму 3х3
```

# 3.Демонстрация атрибутов

```python
# Размер массива и тип данных в нём
two_d_array = np.array([[1, 2.6, 3], [4, 5, 6.0]])
nx_d_array = np.full((2,4,1,5), 7)
str_array = np.array((["tomato", "red bean buns", "melon"], ["Zhu Yuan", "Yanagi", "Miyabi"]))

print(two_d_array.shape, two_d_array.dtype)
print(nx_d_array.shape, nx_d_array.dtype)
print(str_array.shape, str_array.dtype) # строка юникод и максимальная длина 13 символов
```

# 4.Примеры индексирования и срезов

```python
# Получение определённого элемента и подмассива
sl_array = np.arange(58, 90)

print(sl_array[4])
print(sl_array[3:9])
```

# 5.Примеры векторных операций и базовых статистик

```python
first_array = np.array([[1,2,3], [89, 20, 1]])
second_array = np.array([[9, 4.3, 1], [1, 2, 10]])
```

```python
# Поэлементное сложение массивов
summed_array = first_array + second_array
print(summed_array, end="\n\n")

# Поэлементно вычитание массивов
substracted_array = first_array - second_array
print(substracted_array, end="\n\n")

# Поэлементно перемножение массивов
multiplied_array = first_array * second_array
print(multiplied_array, end="\n\n")

# Поэлементно деление массивов
div_array = first_array / second_array
print(div_array)
```

```python
# Среднее значение
print(np.mean(first_array))
print(np.mean(second_array), end="\n\n")

# Медиана
print(np.median(first_array))
print(np.median(second_array), end="\n\n")

# Стандартное отклонение
print(np.std(first_array))
print(np.std(second_array), end="\n\n")

# Дисперсия
print(np.var(first_array))
print(np.var(second_array), end="\n\n")
```
